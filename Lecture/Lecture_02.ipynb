```python
import pandas as pd

df = pd.read_csv('Fish.csv')
print(type(df))
df = df[df['Species'] == 'Perch']
```

```python
x = df[['Length2', 'Height', 'Width']]
y = df[['Weight']]
```

```python
x = x.to_numpy()
y = y.to_numpy()

print(x.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
print(x_test.shape, y_test.shape)
```

### Chapter 02. 사이킷런 변환기

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(include_bias=False) # 변환기 클래스
         # x0, x1
poly.fit([[2, 3], # 첫번째 농어의 길이와 높이
          [4, 5], # 두번째 농어의 길이와 높이
          [6, 7]]) # 세번째 농어의 길이와 높이

# y = a * x0 + b + x1 + c * 1 <----- c: y 절편 = Bias(편향)
# 오리지널 특성: x0, x1
# 특성 공학: x0^2, x1^2, x0 * x1

poly.transform([[2, 3],
                [4, 5], 
                [6, 7]])
```

```python
print(poly.get_feature_names_out())
```

```python
poly = PolynomialFeatures(include_bias=False)

poly.fit(x_train)
# 오리지널: x0, x1, x2
# 특성공학: x0^2, x1^2, x2^2, x0 x1, x0 x2, x1 x2 <--- 3개 + 3개 

x_train_poly = poly.transform(x_train)
print(x_train_poly.shape)
print(x_train_poly)
```

```python
x_test_poly = poly.transform(x_test)
print(x_test_poly.shape)
```
